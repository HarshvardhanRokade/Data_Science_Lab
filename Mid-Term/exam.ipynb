```python
import pandas as pd

df = pd.read_csv("students.csv")

# Display the first 5 rows
print("First 5 rows:")
df.head()
```

```python
print("First 5 rows:")
df.tail()
```

```python
print("Data Info:")
print(df.info())
```

```python
# 3. Summary statistics
print("3. Summary Statistics:")
print(df.describe())
```

```python
# 4. Select specific columns (Name and Score)
print("4. Selected Columns (Name and Score):")
print(df[['Name', 'Score']].head())
```

```python
# 5. Filter rows where Score > 80
print("\n5. Students with Score > 80:")
print(df[df['Score'] > 80].head())
```

```python
# 6. Sort by Score in descending order
print("\n6. Sorted by Score Descending:")
print(df.sort_values(by='Score', ascending=False).head())
```

```python
# 7. Group by Grade and find the mean Score
print("\n7. Mean Score by Grade:")
print(df.groupby('Grade')['Score'].mean())
```

```python
# 8. Group by Subject and get the max score
print("\n8. Max Score by Subject:")
print(df.groupby('Subject')['Score'].max())
```

```python
# 9. Add a new column 'Pass' where Score >= 60 is True else False
df['Pass'] = df['Score'] >= 60
print("\n9. Data with Pass Column:")
print(df.head())
```

```python
# 10. Find the average age
print("\n10. Average Age:")
print(df['Age'].mean())
```

```python
# 11. Count how many students are in each Grade
print("\n11. Count of Students by Grade:")
print(df['Grade'].value_counts())
```

```python
# 12. Apply a function to scale scores by 10%
df['Scaled_Score'] = df['Score'] * 1.1
print("\n12. Data with Scaled Score:")
print(df.head())
```

```python
# 13. Filter students who passed and are in Math
print("\n13. Students who Passed and are in Math:")
print(df[(df['Pass'] == True) & (df['Subject'] == 'Math')].head())
```

```python
#Check for missing values
print(df.isnull().sum())
```

```python
#Check for duplicates
print(df.duplicated().sum())
```

```python
#Fill missing values
df.fillna(0, inplace=True)  
```

```python
#Drop missing values
df.dropna(inplace=True)
```

```python
#Drop duplicate rows
df.drop_duplicates(inplace=True)
```

```python
#Filter with multiple conditions
result = df[(df['Score'] > 10) & (df['Subject'] == 'Science')]
```

```python
result = df.query("Score > 70 and Subject == 'Science'")
```

```python
#Aggregate multiple functions at once
print(df.groupby('Subject')['Score'].agg(['mean', 'min', 'max']))
```

```python
#Pivot tables
pivot = df.pivot_table(values='Score', index='Subject', columns='Grade', aggfunc='mean')
print(pivot)
```

```python
#Rank the scores
df['Rank'] = df['Score'].rank(ascending=False)
print(df.head())
```

```python
#Convert all names to uppercase
df['Name'] = df['Name'].str.upper()
```

```python
#Find rows where Name starts with a letter
print(df[df['Name'].str.startswith('A')])
```

```python
df.groupby(['Grade', 'Subject'])['Score'].mean()
```

```python
#Histogram
df['Score'].hist()
```

```python
#Bar chart
df.groupby('Subject')['Score'].mean().plot(kind='bar')
```
