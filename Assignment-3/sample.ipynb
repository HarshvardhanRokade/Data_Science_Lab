```python
# Program to calculate Mean, Median and Mode without using inbuilt functions

data = [12, 15, 12, 18, 19, 12, 17, 20, 18, 15]

#Mean
def calculate_mean(data):
    return sum(data) / len(data)

#Median
def calculate_median(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    mid = n // 2
    if n % 2 == 0:
        median = (sorted_data[mid - 1] + sorted_data[mid]) / 2
    else:
        median = sorted_data[mid]
    return median

#Mode
def calculate_mode(data):
    freq = {}
    for val in data:
        freq[val] = freq.get(val, 0) + 1
    max_freq = max(freq.values())
    mode = [key for key, value in freq.items() if value == max_freq]
    return mode

print("\nWithout Using Inbuilt Functions:")
print("Mean:", calculate_mean(data))
print("Median:", calculate_median(data))
print("Mode:", calculate_mode(data))
```

```python
# Program to calculate Mean, Median and Mode using inbuilt functions

import statistics as stats

data = [12, 15, 12, 18, 19, 12, 17, 20, 18, 15]

# Mean
mean_value = stats.mean(data)

# Median
median_value = stats.median(data)

# Mode
mode_value = stats.mode(data)

print("Using Inbuilt Functions:")
print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

```python

import pandas as pd
import statistics as stats

# Create sample CSV data for demonstration
csv_data = '''Student,Marks
S1,78
S2,65
S3,80
S4,92
S5,85
S6,75
S7,88
S8,90
S9,70
S10,85
'''
with open('student_marks.csv', 'w') as f:
    f.write(csv_data)

# Read the CSV
df = pd.read_csv('student_marks.csv')
data = df['Marks'].tolist()
print("Marks List:", data)

# Inbuilt calculations
mean_inbuilt = stats.mean(data)
median_inbuilt = stats.median(data)
mode_inbuilt = stats.mode(data)

# Without inbuilt functions
def calculate_mean(data):
    return sum(data) / len(data)

def calculate_median(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    mid = n // 2
    if n % 2 == 0:
        return (sorted_data[mid - 1] + sorted_data[mid]) / 2
    else:
        return sorted_data[mid]

def calculate_mode(data):
    freq = {}
    for num in data:
        freq[num] = freq.get(num, 0) + 1
    max_count = max(freq.values())
    modes = [k for k, v in freq.items() if v == max_count]
    return modes[0] if len(modes) == 1 else modes

mean_manual = calculate_mean(data)
median_manual = calculate_median(data)
mode_manual = calculate_mode(data)

print("\nExpected Output:\n")
print(f"Mean using inbuilt function: {mean_inbuilt}")
print(f"Median using inbuilt function: {median_inbuilt}")
print(f"Mode using inbuilt function: {mode_inbuilt}")
print(f"Mean without inbuilt function: {mean_manual}")
print(f"Median without inbuilt function: {median_manual}")
print(f"Mode without inbuilt function: {mode_manual}")
```
