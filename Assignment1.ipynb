```python
import numpy as np

array2 = np.array([[1,2,3,4],[5,6,7,8]])            #creating two dimension array
array1 = np.array([1,2,3,4])                        #creating an array
array1                                              #printing an array
array1.dtype                                        #printing the datatype
array1.ndim                                         #printing the dimension
array2.ndim
array2.shape                                        #numbers of rows and columns
array1.reshape(2,2)                                 #convert dimension of an array
print(array1[0])                                    #array indexing
print(array1[-1])                                   #to access last element of array
array3 = np.arange(0,11)                            #printing array using arange function
print(array3)  
print(np.argmax(array1))                            #index of maximum element in array
print(np.argmin(array2))                            #index of minimum element in array
print(np.diff(array1))                              #difference between consecutive elements
print(np.linspace(0,1,6))                           #evenly spaced values between two numbers
```

```python
x = np.array([1,2,3])
y = np.array([4,5,6])
add = x + y                                         #add elements of array
sub = x - y                                         #subtract elements of array
mul = x * y                                         #multiple elements of array
div = x / y                                         #divide elements of array
print(add)
print(sub)
print(mul)
print(div)
```
