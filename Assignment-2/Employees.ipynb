```python
import pandas as pd

df = pd.read_csv("employee_dataset.csv")
df.head()
```

```python
df.isnull().sum()
df.fillna({'Salary': df['Salary'].mean()}, inplace=True)
df.fillna({'WorkHours': df['WorkHours'].mean()}, inplace=True) 
df.fillna({0: 0,  'PerformanceScore': df['PerformanceScore'].mean()},inplace=True)
```

```python
df.groupby('Department')['Salary'].mean()             #Average salary by department
```

```python
df.groupby('Gender')['Salary'].mean()                   #Average salary by gender
```

```python
df.groupby('Department')[['Salary', 'PerformanceScore']].mean()         #Average salary & performance score by department
```

```python
df.groupby('Department')['EmpID'].count()                   #Count employees in each department
```

```python
df.groupby(['Department', 'Gender'])['Salary'].mean()           #Group by multiple columns
```

```python
df['Salary'].plot(kind='hist')     # histogram
```

```python
df['WorkHours'].plot(kind='box')      # boxplot
```

```python
df.plot(x='Department', y='PerformanceScore', kind='scatter')  # scatter plot
```
