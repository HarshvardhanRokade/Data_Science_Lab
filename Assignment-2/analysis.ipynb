```python
# Name : Harshvardhan Rokade
# Roll no : 117
# Division : B




import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'Product': ['Laptop', 'Smartphone', 'Headphones', 'Tablet', 'Smartwatch', 'Camera'],
    'Category': ['Computers', 'Mobiles', 'Accessories', 'Computers', 'Wearables', 'Photography'],
    'Brand': ['Dell', 'Samsung', 'Sony', 'Apple', 'Fitbit', 'Canon'],
    'Price': [75000, 55000, 5000, 30000, 15000, 40000],
    'Units_Sold': [120, 250, 400, 180, 150, 90],
    'Customer_Rating': [4.5, 4.2, 4.0, 4.6, 4.1, 4.3]
}

df = pd.DataFrame(data)
df
```

```python
# Qualitative Analysis
print("Qualitative Analysis:")
print(df['Category'].value_counts())
print(df['Brand'].value_counts())

# Bar chart: Brand vs Units Sold
plt.figure(figsize=(7,5))
plt.bar(df['Brand'], df['Units_Sold'], color=plt.cm.Paired.colors, edgecolor='black')
plt.title("Units Sold by Brand")
plt.xlabel("Brand")
plt.ylabel("Units Sold")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
# Quantitative Analysis
print("\nQuantitative Analysis:")
print(df[['Price', 'Units_Sold']].describe())

# Histogram: Customer Rating distribution
plt.figure(figsize=(7,5))
plt.hist(df['Customer_Rating'], bins=5, color='teal', edgecolor='white', alpha=0.8)
plt.title("Customer Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
# Scatter plot: Units Sold vs Price (swapped axes)
plt.figure(figsize=(7,5))
scatter = plt.scatter(
    df['Units_Sold'], df['Price'],
    c=df['Customer_Rating'], cmap='viridis', s=100, edgecolor='black'
)
plt.colorbar(scatter, label="Customer Rating")
plt.title("Units Sold vs Price")
plt.xlabel("Units Sold")
plt.ylabel("Price")

# Adding labels
for i, product in enumerate(df['Product']):
    plt.annotate(product, (df['Units_Sold'][i], df['Price'][i]), fontsize=8, xytext=(5,5), textcoords='offset points')

plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
